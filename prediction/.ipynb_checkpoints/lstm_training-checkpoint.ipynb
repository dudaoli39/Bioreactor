```python
import datetime
import sklearn
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import KernelPCA
import numpy as np
import pandas as pd
import math
import keras
import matplotlib.pyplot as plt
import tensorflow as tf
tf.random.set_seed(99)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense
```

READ THE EXTRACTED MERGED FILE FROM THE MONGODB COLLECTIONS

```python
# Dataset loading
dset_address = 'datasets\merged.csv'
dataFrame = pd.read_csv(dset_address)
```

PRE PROCESSING THE VALUES TO PREPARE FOR TRAINING

```python
#working on missing values (if any) and removing the DATE column as it is not necessary for training
imputer = SimpleImputer(missing_values=np.nan)  # Handling missing values
dataFrame.drop(columns=['date'], inplace=True)
dataFrame = pd.DataFrame(imputer.fit_transform(dataFrame), columns=dataFrame.columns)
dataFrame = dataFrame.reset_index(drop=True)
```

```python
# Applying feature scaling on all columns to fit in the desired interval for comparison
scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = scaler.fit_transform(dataFrame.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=list(dataFrame.columns))
target_scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled[['temperature', 'pressure', 'humidity', 'oxigen']] = target_scaler.fit_transform(dataFrame[['temperature', 'pressure', 'humidity', 'oxigen']].to_numpy())
df_scaled = df_scaled.astype(float)
```

DATA PREPARATION

```python
# Single step dataset preparation
# it will read all values and encapsulate it into inputs and desired target values
def singleStepSampler(df, window):
    xRes = []
    yRes = []
    for i in range(0, len(df) - window):
        res = []
        for j in range(0, window):
            r = []
            for col in df.columns:
                r.append(df[col][i + j])
            res.append(r)
        xRes.append(res)
        yRes.append(df[['temperature', 'pressure', 'humidity', 'oxigen']].iloc[i + window].values)
    return np.array(xRes), np.array(yRes)
```

DATA SPLIT INTO TRAINING AND TEST

```python
# Dataset splitting
SPLIT = 0.8
(xVal, yVal) = singleStepSampler(df_scaled, 20)
X_train = xVal[:int(SPLIT * len(xVal))]
y_train = yVal[:int(SPLIT * len(yVal))]
X_test = xVal[int(SPLIT * len(xVal)):]
y_test = yVal[int(SPLIT * len(yVal)):]
```

```python
print(len(X_train), len(X_test))
```

LSTM MODEL CREATION

```python
# Initialize the model
model = Sequential()
```

```python
# Add an LSTM layer with 200 units and 4 variables to predict
model.add(LSTM(200, return_sequences=False, input_shape=(len(X_train), 4)))

# Add a Dropout layer with a dropout rate of 0.2
model.add(Dropout(0.2))

# Add a Dense layer with 4 units related to the number of classes (for classification)
model.add(Dense(4))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

FORECASTING

```python
# Reload the data with the date index
dataFrame = pd.read_csv(dset_address)
dataFrame['date'] = pd.to_datetime(dataFrame['date'])
dataFrame.set_index('date', inplace=True)
```

```python
# Forecast Plot with Dates on X-axis
predicted_values = model.predict(X_test)
```

```python
d = {
    'Predict_Temp': predicted_values[:, 0],
    'Predict_Pres': predicted_values[:, 1],
    'Predict_Humi': predicted_values[:, 2],
    'Predict_Oxig': predicted_values[:, 3],
    'Actual_Temp': y_test[:, 0],
    'Actual_Pres': y_test[:, 1],
    'Actual_Humi': y_test[:, 2],
    'Actual_Oxig': y_test[:, 3],
}

d = pd.DataFrame(d)
d.index = dataFrame.index[-len(y_test):]  # Assigning the correct date index
```

```python
# FUNCTION TO PLOT RESULTS - IT CAN BE USED TO PLOT MORE THAN ONE SENSOR AT A TIME
def plot_results(actual, predict, graphic_name):
    fig, ax = plt.subplots(figsize=(15, 9))
    #  highlight the  forecast
    highlight_start = int(len(d) * 0.9)  
    highlight_end = len(d) - 1  # Adjusted to stay within bounds
    # Plot the actual values
    plt.plot(d[actual][:highlight_start], label=actual)
    
    # Plot predicted values with a dashed line
    plt.plot(d[predict], label=predict, linestyle='--')
    
    # Highlight the forecasted portion with a different color
    plt.axvspan(d.index[highlight_start], d.index[highlight_end], facecolor='lightgreen', alpha=0.5, label='Forecast')
    
    plt.title('BioReactor '+graphic_name+' Time-Series forecasting using LSTM')
    plt.xlabel('Dates')
    plt.ylabel('Values')
    ax.legend()
    plt.show()
```

```python
actual_list = ['Actual_Temp', 'Actual_Pres', 'Actual_Humi', 'Actual_Oxig']
predict_list = ['Predict_Temp', 'Predict_Pres', 'Predict_Humi', 'Predict_Oxig']
plot_results(actual_list, predict_list, "")
```
